```python
import tweepy
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import json
import pandas as pd
from pprint import pprint
```

```python
#Variables that contains the user credentials to access Twitter API 
access_token = ""
access_token_secret = ""
consumer_key = ""
consumer_secret = ""

auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

tweetdata = []

def realTimeTweets():
    class StdOutListener(StreamListener):
        
        def __init__(self, api=None):
            super(StdOutListener, self).__init__()
            self.num_tweets = 0

        def on_data(self, data):
            self.num_tweets +=1
            if self.num_tweets < 500:
                tweets = json.loads(data)
                try:
                    #print(self.num_tweets , '\n')
                    tweetdata.append(tweets['text'])
                except:
                    pass
                return True
            else:
                return False
                
            

        def on_error(self, status):
            print('error' + str(status))

    l = StdOutListener()
    stream = Stream(auth, l)
    stream.filter(languages=["en"], track=['college', 'teenager', 'university'])
    

    
if __name__ == '__main__':
    realTimeTweets()
```

```python
filtereddata = []

#Filter the tweets only English
for line in tweetdata:
    try:
        line.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        pass
    else:
        filtereddata.append(line)
```

```python
testdata = []
count = 0

#Limit the tweets to 100
for line in filtereddata:
    if count < 100:
        print(count)
        print(line)
        testdata.append(line)
        count +=1
        
df = pd.DataFrame(testdata, columns=['content'])
print(df)
df.to_csv('Resources/testing_dataset.csv', index = False)
```
